```python
import numpy as np
import pandas as pd
import math

array = [[-0.2,0.5,0],
        [0.2,-0.5,0],
        [0.8,-0.8,1],
        [0.8,0.8,1]]

w = [(i,j,k) 
     for i in np.arange(-10, 10, 0.1) 
     for j in np.arange(-10, 10, 0.1)
     for k in np.arange(-10, 10, 0.1)]
```

## SLENKSTINE

```python
def threshold_find_valid_for_row(row,w):
    valid = []
    for i in w:
        a = row[0]*i[0] + row[1]*i[1] + i[2]
        if (row[-1] == 0 and a < 0) or (row[-1] == 1 and a >= 0):
            valid.append(i)
    return valid
```

```python
tvalid = w
tvalid_temp = []
for row in array:
    tvalid_temp = threshold_find_valid_for_row(row,tvalid)
    tvalid = tvalid_temp

tvalid = [(round(i,2),round(j,2),round(k,2)) for (i,j,k) in tvalid]
```

```python
print("Pirmas:",tvalid[0],"\nPaskutinis:",tvalid[-1])
```

## SIGMOIDINE
Slenkstis klasei <b>0</b>: f(a) < 0.07<br>
Slenkstis klasei <b>1</b>: f(a) > 0.93

```python
def sigmoid_find_valid_for_row(row,w):
    valid = []
    for i in w:
        a = row[0]*i[0] + row[1]*i[1] + i[2]
        fa = 1/(1+math.exp(-a))
        
        if fa > 0.93:
            fa = 1
        if fa < 0.07:
            fa = 0
        if fa == row[-1]:
            valid.append(i)
    return valid
```

```python
svalid = w
svalid_temp = []
for row in array:
    svalid_temp = sigmoid_find_valid_for_row(row,svalid)
    svalid = svalid_temp
    
svalid = [(round(i,2),round(j,2),round(k,2)) for (i,j,k) in svalid]
```

```python
print("Pirmas:",svalid[0],"\nPaskutinis:",svalid[-1])
```
